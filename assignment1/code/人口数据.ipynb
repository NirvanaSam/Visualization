```python
import csv #处理csv文件的模块
import matplotlib.pyplot as plt #把pyplot命名为别名plt
filename = "../data/world-population.csv" #人口数据的文件
datax = []
datay = []
with open(filename) as f: #打开文件
    reader = csv.reader(f) #读取文件内容
    for datarow in reader: #遍历文件内容
        if reader.line_num != 1:
#             print(reader.line_num,datarow)
            datax.append(datarow[0]) #把时间数据保存到datax列表
            datay.append(datarow[1]) #把人口数据保存到datay列表
add = list(range(0, len(datax),10))
datax = [datax[i] for i in add]
datay = [datay[i] for i in add]
plt.figure(dpi=500)
plt.plot(datax,datay) #画图
plt.show() #显示图像
```
