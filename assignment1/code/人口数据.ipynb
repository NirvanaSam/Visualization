```python
import csv #csv
import matplotlib.pyplot as plt #pyplotplt
filename = "../data/world-population.csv" #
datax = []
datay = []
with open(filename) as f: #
    reader = csv.reader(f) #
    for datarow in reader: #
        if reader.line_num != 1:
#             print(reader.line_num,datarow)
            datax.append(datarow[0]) #datax
            datay.append(datarow[1]) #datay
add = list(range(0, len(datax),10))
datax = [datax[i] for i in add]
datay = [datay[i] for i in add]
plt.figure(dpi=500)
plt.plot(datax,datay) #
plt.show() #
```
